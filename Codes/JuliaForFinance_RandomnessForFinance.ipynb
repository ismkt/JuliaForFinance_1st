# Juliaで動かす金融計算-乱数を使う話
ここからは, 実際に金融の計算...の前に, 金融の計算で大事になる乱数の取り扱いを練習したいと思います.  
株式や金利, スワップレートなど, 金融の世界では時々刻々, その値が変化するものが多くあります.  
それらの値が将来どうなるかを考えてみるのは, 得られる利益を予測したり, あるいはリスク計算をする上でとても重要です.  
しかし, 将来どういった値をとるかというのは1点読みできるようなものではありません.  
そこで, 乱数を使って, 将来の値を確率的に（=ランダムに）生成することによって, シミュレーションしてみようという試みがあります.  
ここでは, そんな株式や金利などの値をシミュレーションするための前段階として, 乱数の使い方に慣れておきましょう.

## 乱数の作り方


そもそも乱数とは何かというと, 簡単に言えば実行される都度値の変わるもの, といったところでしょうか.  
`Julia`における乱数の生成の仕方は標準ライブラリを利用する方法もありますが, 今後の内容を見据えて, `Distributions`というパッケージを使います.  
`Distributions`は様々な確率分布（=どの値がどれくらい出やすいかを示すもの）を扱うパッケージです.  
まずは, `Distributions`を使って, 一様分布からの乱数生成方法を見ていきましょう.

### 一様分布


一様分布（Uniform Distribution）というのは, 確率分布の一種で, ある範囲の値が偏りなく出るようなものを言います.  
例としては, サイコロを1回投げたときに出る目が挙げられます. 出る目は1～6のどれかで, それらの出る確率は等しいです. この時, 「サイコロの出る目は（離散）一様分布に従う」なんて言ったりします.  
堅苦しい表現ですが, 言いたいことは結局 **「一様分布⇒どれも同じ確率で起こるよ」** というだけです.  
また, 離散という, なんとも堅苦しい言葉が付いていることに嫌悪感を抱く方がいるかもしれません.  
ただ言ってしまえば, **「離散⇒飛び飛びの値だよ」** くらいの意味です. 先ほどのサイコロの例だと, 出る目が1, 2, 3 , ... , 6であって, 1.1とか0.5とかそういった値が出ることなく, 1刻みでしか値が出ませんよね？ こういった意味でサイコロの出る目は離散的（=飛び飛びの値）なのです.  
では逆に, **離散的でないものは「連続的」** といいます. 例えばExcelで`RAND()`という関数を使うと, 0～1までの連続一様分布に従う（つまり, 0.01も0.5も0.31231もその他0～1までのどんな値も, 全て同じ確率で出現する）乱数を出力できます.  
では, いい加減にコーディングに移っていきましょう.  

```julia
using Distributions #使うパッケージを宣言

#離散一様分布から乱数を作る
function MakeRandomFromDiscreteUniform(n) #Discrete = 離散
    d = DiscreteUniform(1, 6) #1～6までの離散一様分布を生成（要するにサイコロ）
    res = rand(d, n) #分布からn個の乱数を生成
    println(res) #生成した乱数を表示
end

MakeRandomFromDiscreteUniform(5)
```

```julia
using Distributions #使うパッケージを宣言

#連続一様分布から乱数を作る
function MakeRandomsContinuousUniform(n)
    d = Uniform(0 , 1) #0から1までの連続一様分布を生成
    res = rand(d, n) #分布からn個の乱数を生成
    println(res) #生成した乱数を表示
end

MakeRandomsContinuousUniform(5)
```

どうでしょうか？  
出力される結果は, 実行する毎に異なるはずです. 何度か実行して遊んでみてください.  
さて, コードを見て, 鋭い人は気づくかもしれませんが, 実は二つの関数`DiscreteuniformRandoms`と`ContinousUniformRandoms`はほとんど同じコードです.  
これは`Julia`の`Distributions`のお陰なのです. `Distributions`の`rand(d, n)`の第一引数である`d`は確率分布なら何でもいいので, 確率分布さえ作ってしまえば, そこからの乱数生成は勝手にやってくれるのです.  
しかも, 大抵の確率分布は`Distributions`に備え付けで入っているので, 自分で1から確率分布を頑張って作る必要はほとんどありません.  

せっかく確率分布を使っているのだから, どんな値がどれくらい出やすいかを, グラフで見てみましょう.  
これは`Distributions`と, `StatsPlots`の二つのパッケージの合わせ技で実現します.  
先ほどの関数をちょっと改造していきましょう.  

```julia
using Distributions #使うパッケージを宣言
using StatsPlots #パッケージを追加

#離散一様分布から乱数を作る
function MakeRandomsFromDiscreteUniform(n) #Discrete = 離散
    d = DiscreteUniform(1, 6) #1～6までの離散一様分布を生成（要するにサイコロ）
    res = rand(d, n) #分布からn個の乱数を生成
    println(res) #生成した乱数を表示
    plot(d) #確率分布をグラフに表示
end

MakeRandomsFromDiscreteUniform(5)
```

```julia
using Distributions #使うパッケージを宣言
using StatsPlots #パッケージを追加

#連続一様分布から乱数を作る
function MakeRamdomsFromContinuousUniform(n)
    d = Uniform(0 , 1) #0から1までの連続一様分布を生成
    res = rand(d, n) #分布からn個の乱数を生成
    println(res) #生成した乱数を表示
    plot(d, ylims = (0.0, 1.1)) #確率分布をグラフに表示
end

MakeRamdomsFromContinuousUniform(5)
```

確率分布がグラフに表示されたと思います.  
**横軸がどんな値が出るかで, 縦軸の値がどれくらい出やすいか** (※)を表しています.  
一様分布を眺めても面白くないかもしれませんが, どの値がどれくらい出やすいかをイメージしておくのはとても大事なので, グラフが書けるときは書いておきましょう.

### 正規分布

一様分布では, どの値も偏りなく出る確率分布を扱いました.  
もちろん, 一様分布も大事なのですが, 金融の世界ではもっと重要な確率分布があります.  
それは ***正規分布*** です.  
正規分布は, 平均値あたりが出やすくて, 平均から離れた値はあまり出ないような確率分布です.  
（連続）一様分布では出る値の範囲を決めるために2つの変数を使いましたが, 正規分布では **平均値(mean)** と, 平均から離れた値の出やすさを決める **分散(variance)** の2つの変数を使います.  
それでは, コードを書いて乱数の生成と確率分布の形を見てみましょう.  

```julia
using Distributions
using StatsPlots

function MakeRandomsFromNormalDistribuion(n)
    d = Normal(0, 1^0.5) #平均が0で標準偏差(=分散の0.5乗)が1の正規分布を生成
    res = rand(d, n)
    println(res)

    plot(d)
end

MakeRandomsFromNormalDistribuion(5)
```

確率分布の形が釣り鐘型になっているのがわかると思います.  
分散が大きくなると, この釣り鐘が横に広がります.  
なぜ, この正規分布が大事かという説明は省力しますが, 金融の世界でよく出てくる分布だということは知っておいて損はありません.

## `rand()`のちょっとした補足

`rand(d, n)`で確率分布から乱数を生成するときに, 1個だけ乱数を生成する場合は`rand(d)`でもOKです.  
これはこれで覚えておきましょう.

## 乱数のまとめ

今までのことをまとめておきましょう.

* 確率分布⇒どんな値がどれだけ出やすいか
* `Distributions`⇒確率分布を簡単に使えるパッケージ
* `Plots`, `StatsPlots`⇒グラフが簡単に書けるパッケージ
* `Julia`での乱数の作り方
    1. 確率分布`d`を作る
    1. 何個乱数を作るか(=`n`)決める
    1. `rand(d, n)`で乱数生成
* 一様分布⇒どの値も, 出る確率が等しい
* 正規分布⇒平均値あたりが出やすい
