## ドットシンタックス


`Julia`には, 配列の各要素の関数を適用する際に便利な記法として, `.`（ドットシンタックス）と呼ばれるものがあります.  
他の言語では`map`や`filter`何て名前で, 関数として実装されています.（もちろん`Julia`にも`map`関数はあります.）  
早速, その書き方を見てみましょう.

```julia
function dot_syntax()
    arr = Int[1, 2, 3]

    println(arr .* 2) #各要素を2倍した結果を表示
    
    #値を3倍する関数を作る
    f = (a) -> begin
        return a * 3
    end

    println(f.(arr)) #各要素にf()を適用して結果を表示

    println(arr) #元の配列の中身は変わらない
end

dot_syntax()
```

このように, `.`という記法を使うことで, 1行で配列の各要素に対する操作結果を得ることができます.  
ちなみに, `+`や`-`, `*`などの`a * b`の形で使うものを2項演算子と言います.  
`arr .* 2`の形で, `.`の右側に付けられるのは2項演算子のみなので注意してください.

## 文字列結合


`Julia`における文字列結合は, ちょっと他の言語と異なります.  
`Julia`では文字列結合に`*`を使います.  
`+`ではないことに注意してください.  
「なぜ掛け算の記号である`*`を文字列結合に使うか?」ということに関しては, 実は数学的な思想があるのですが, 気になる人は自分で調べてみてください.  
では, 実際に文字列結合をしてみましょう.

```julia
function concat()
    s1 = "Hello"
    s2 = "World!"

    s = s1 * " " * s2 #*を使って文字列結合
    println(s)
end

concat()
```

## 関数名と破壊的な操作


`Julia`では関数名にちょっとした慣習があります.  
それは, 破壊的な操作をする関数には`!`を関数名の末尾につけるというものです.  
破壊的な操作とは, 引数で受け取ったものの中身を変えてしまうような操作を言います.  
これは, 具体例でお話したほうがいいでしょうから, 実際にコードをお見せします.  

```julia
function destructive_method_demo()
    a = [1, 3, 2]
    b = [1, 3, 2]

    #sort()⇒配列を並び替えたものを返す関数
    res1 = sort!(a) #引数で受け取ったaの中身を書き換えてしまう関数
    res2 = sort(b) #引数で受け取ったbの中身は書き換わらない.

    println("a:", a)
    println("res1:", res1)

    println("b:", b)
    println("res2:", res2)
end

destructive_method_demo()
```

このように`!`を付けた`sort!()`は引数で受け取った配列の中身を書き換えてしまうのです.  
一方`!`のない`sort()`は引数で受け取った配列の中身は書き換えません.  
つまり, `!`がつくと破壊的な操作をする関数で, `!`がつかないと非破壊的な操作になります.  
自分で関数を作った際や, パッケージを利用する際に覚えておいてください.

### Distributions


`Distributions`は乱数や, 確率分布と呼ばれるものを簡単に扱うためのパッケージです.  
金融計算では, 様々な箇所で確率に関する計算が必要になるため, 結構重宝します.  
本講座では, あまり数学的なことには立ち入らずに進めていく予定ですが, 興味がある人は自分で, 公式サイトで使い方を確認しておくといいでしょう.  
[公式サイト](https://juliastats.org/Distributions.jl/stable/) 

```julia
using Distributions

function random_sampling(n)
    d = Normal(2, 4) #平均が2, 標準偏差が4の正規分布を生成
    randoms = rand(d, n) #正規分布からn個の乱数を生成
    println(randoms)
end

random_sampling(5)#5個の乱数を生成
```

### Plots, StatsPlots


`Plots`及び`StatsPlots`はどちらもグラフ描画用のライブラリです.  
`StatsPlots`の方は, 確率や統計に絡むプロット（ヒストグラムのプロットや確率分布のプロット等）に対応しているイメージです.  
こちらもここでは簡単な使い方のみ紹介しますが, 興味があれば, 自分で公式サイトで使い方を見てみるとよいでしょう.

[公式サイト（Plots）](https://docs.juliaplots.org/stable/)  
[公式サイト（StatsPlots）](https://docs.juliaplots.org/latest/generated/statsplots/)

```julia
using Plots
using StatsPlots

function plots_package_demo()
    #Plotsのデモ
    x = 0:0.01:10 #0から10まで0.01刻みで生成
    y = sin.(x) #xの各要素に関数を適用した結果を格納

    p1 = plot(x, y) #x, yのグラフ描画

    #StatsPlotsのデモ
    randoms = randn(1000) #乱数を生成
    p2 = histogram(randoms) #ヒストグラムの描画

    plot(p1, p2, layout=(2, 1)) #作ったグラフをプロット
end

plots_package_demo()
```
